```python
import os
import time
from urllib.parse import urlparse
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.keys import Keys

def scroll_to_bottom(driver, element_selector):
    element = driver.find_element_by_css_selector(element_selector)
    actions = webdriver.ActionChains(driver)
    
    for _ in range(100):  # Adjust the range value based on the content length
        actions.move_to_element(element)
        actions.click()
        actions.send_keys(Keys.END)
        actions.perform()
        time.sleep(0.5)

NUM_COUNTRIES = 137
```

```python
urls = []

for n in range(1,NUM_COUNTRIES+1):
    url = 'https://global.essentialmeds.org/dashboard/countries/' + str(n)
    urls.append(url)
```

```python
urls
```

```python
output_directory = 'HTML Files New'

if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# Set up headless Firefox options
options = Options()
options.add_argument('-headless')

# Initialize the Firefox WebDriver
driver = webdriver.Firefox(options=options)

for url in urls:
    driver.get(url)

    # Optional: Wait for the JavaScript to load
    import time
    time.sleep(5)

    scroll_to_bottom(driver, '.tbody.ReactVirtualized__Grid__innerScrollContainer')  # Replace 'your_element_selector' with the actual selector

    soup = BeautifulSoup(driver.page_source, 'html.parser')

    parsed_url = urlparse(url)
    domain = parsed_url.netloc
    path = parsed_url.path.strip('/').replace('/', '_')

    if path == '':
        path = 'index'

    filename = os.path.join(output_directory, f'{domain}_{path}.html')

    with open(filename, 'w', encoding='utf-8') as file:
        file.write(str(soup.prettify()))

# Close the browser
driver.quit()
```
